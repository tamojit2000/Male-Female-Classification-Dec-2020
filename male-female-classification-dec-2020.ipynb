```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2

from tensorflow.keras.preprocessing import image
```

```python
men = []
women = []
img_size = 224
MEN_IMGS_PATH = '../input/men-women-classification/men'
WOMEN_IMGS_PATH = '../input/men-women-classification/women'
DIRS = [(1, MEN_IMGS_PATH), (0, WOMEN_IMGS_PATH)]
```

```python
train_images = []
labels = []
for num, _dir in DIRS:
    _dir = _dir + '/'
    count = 0
    for file in os.listdir(_dir):
        if count >= 1400:
            break
        img = image.load_img(_dir + str(file), target_size=(img_size, img_size))
        img = image.img_to_array(img)
        img=img/225.0
        train_images.append(img)
        labels.append(num)
        count += 1
        print('\r', count, end='',flush=True)
```

```python
X = np.array(train_images)
```

```python
len(labels)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle

#X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.1, random_state=101)
X,Y=shuffle(X,labels)
X=np.array(X)
Y=np.array(Y)
```

```python
import tensorflow as tf

class myCallBack(tf.keras.callbacks.Callback):
    def on_epoch_end(self,epoch,logs={}):
        if(logs.get('accuracy')-logs.get('val_accuracy') > 0.05) or (logs.get('accuracy')>0.96) or (logs.get('val_accuracy')>0.96):
            self.model.stop_training = True

callback=myCallBack()

image_size=224

base_model = tf.keras.applications.MobileNet(weights="imagenet",include_top=False,input_shape=(image_size,image_size,3))
base_model.trainable = False


x = base_model.output
x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = tf.keras.layers.Dropout(0.25)(x)
x = tf.keras.layers.Dense(8,activation='relu')(x)
x = tf.keras.layers.Dense(2,activation="softmax")(x)
model = tf.keras.Model(inputs=base_model.input, outputs=x)
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])

history=model.fit(X, Y,epochs=50,validation_split=0.25,callbacks=[callback])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
from sklearn.metrics import confusion_matrix,classification_report
predictions = np.argmax(model.predict(X),axis=1)
print(confusion_matrix(Y, predictions))
print(classification_report(Y,predictions))
```
